```python
# Computing statistics per student
student_stats = df.groupby('user_id').agg(
    problems_practiced=('problem_id', 'nunique'),
    avg_attempts=('attempt_count', 'mean'),
    accuracy=('correct', 'mean')
).reset_index()
```

```python
# Problems practiced vs. Accuracy
plt.figure(figsize=(8, 5))
sns.scatterplot(data=student_stats, x='problems_practiced', y='accuracy', alpha=0.6)
sns.regplot(data=student_stats, x='problems_practiced', y='accuracy', scatter=False, color='red')
plt.title('Accuracy vs. Number of unique problems practiced')
plt.xlabel('Unique problems practiced')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
#This graph shows that more the problems practiced better is the accuracy
```

```python
# Average attempts per problem vs. Accuracy
plt.figure(figsize=(8, 5))
sns.scatterplot(data=student_stats, x='avg_attempts', y='accuracy', alpha=0.6)
sns.regplot(data=student_stats, x='avg_attempts', y='accuracy', scatter=False, color='green')
plt.title('Accuracy vs. Average attempts per roblem')
plt.xlabel('Average attempts')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
#accuracy of the user is shown to decrease when attempting a problem too many times
```

```python
# Plot 3: Problems practiced vs. Average attempts (to check persistence)
plt.figure(figsize=(8, 5))
sns.scatterplot(data=student_stats, x='problems_practiced', y='avg_attempts', alpha=0.6)
sns.regplot(data=student_stats, x='problems_practiced', y='avg_attempts', scatter=False, color='blue')
plt.title('Avg attempts vs. Number of problems practiced')
plt.xlabel('Unique problems practiced')
plt.ylabel('Average attempts')
plt.grid(True)
plt.show()
#from the graph we will see that the avg attempts decrease as the more problems are practiced
```
